#Take content (Lấy văn bản gốc)

```python
def take_content(container):
  take_link = container.find('a').attrs['href']
  my_url2 = str(take_link)
  uclient2 = Ureq(my_url2)
  page_html2 = uclient2.read()
  uclient2.close()
  page_soup2 = soup(page_html2,'html.parser')

  content =''
  c=0
  containers2__ = page_soup2.find('div',{'itemprop': 'articleBody', 'class': 'detail_content', 'id': 'detail_ct'})
  containers2 = containers2__.find_all('p')
  # print(len(containers2))
  for x in containers2:
    c +=1
    if x.text == '&nbsp;':
      continue
    if c == len(containers2):
      continue
    content += (x.text) +'\n'

  return content
```

```python
def take_content(container):
  take_link = container.find('a').attrs['href']
  my_url2 = str(take_link)
  uclient2 = Ureq(my_url2)
  page_html2 = uclient2.read()
  uclient2.close()
  page_soup2 = soup(page_html2,'html.parser')

  content =''
  c=0
  containers2__ = page_soup2.find('div',{'itemprop': 'articleBody', 'class': 'detail_content', 'id': 'detail_ct'})
  containers2 = containers2__.find_all('p')
  # print(len(containers2))
  for x in containers2:
    c +=1
    if x.text == '&nbsp;':
      continue
    if c == len(containers2):
      continue
    content += (x.text) +'\n'

  return content
```

```python
listofdicts=[]
count = 1
data=[]
headers=['title', 'content', 'abstract'] 
for i in range(1,101): # 1,251 # có trùng lặp kiều báo lát xóa
  page_web =str(i)
  my_url = 'https://docbao.vn/doi-song/p'+page_web
  uclient = Ureq(my_url)
  page_html = uclient.read()

  uclient.close()
  page_soup = soup(page_html,'html.parser')

  containers__ = page_soup.find('div', {'class':'category_news_list'})
  containers_ = containers__.find('ul')
  containers = containers_.find_all('li')

  for container in containers:
    lst =[]
    count+=1
    title = container.find('h3').find('a').text
    try:
      Content = take_content(container)
    except:
      print('*'*30 + 'CHÚ Ý')
      print('title: ',title)
      print(page_web)
      pass
    s_content = container.find('p').text
    
    lst.append(title)
    lst.append(Content)
    lst.append(s_content)
    data.append(lst)
    print('Title: '+title)
    print('Content: ',Content)
    print('Abstract: '+s_content)
    print(count)
  print(page_web)
```
